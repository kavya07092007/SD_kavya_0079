```python
def fibonacci(n):
  a=0
  b=1
  for i in range(n):
    print(a,end=" ")
    c=a+b
    a=b
    b=c
n=int(input("enter the number of terms"))
fibonacci(n)

```

```python
def test_fib(n):
  a,b=0,1
  for i in range(n):
    yield a
    a,b = b, a+b
```

```python
for i in test_fib(10):
  print(i)
```

```python
def test_fib1():
  a,b=0,1
  while True:
    yield a
    a,b=b , a+b
```

```python
fib=test_fib1()
```
